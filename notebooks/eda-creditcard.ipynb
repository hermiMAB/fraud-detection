```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style="whitegrid")

# --- 1. Load Data ---
creditcard_df = pd.read_csv('../data/raw/creditcard.csv')
print(f"Initial Credit Card Data Shape: {creditcard_df.shape}")

# --- 2. Data Cleaning ---
# Drop missing values and duplicates
creditcard_df.dropna(inplace=True)
creditcard_df.drop_duplicates(inplace=True)
print(f"Cleaned Credit Card Data Shape: {creditcard_df.shape}\n")

# --- 3. Quantify Class Imbalance ---
print("--- Credit Card Data Class Distribution ---")
print(creditcard_df['Class'].value_counts(normalize=True) * 100)
```

```python

# --- 4. Univariate Analysis ---
plt.figure(figsize=(10, 5))
# Filtering extremely high outlier amounts just for better visualization
sns.histplot(creditcard_df[creditcard_df['Amount'] < 500]['Amount'], bins=50, kde=True, color='green')
plt.title("Univariate: Distribution of Transaction Amounts (Bank Data)")
plt.xlabel("Transaction Amount ($)")
plt.ylabel("Frequency")
plt.show()

```

```python
# --- 5. Bivariate Analysis ---
plt.figure(figsize=(10, 5))
sns.boxplot(data=creditcard_df, x='Class', y='Amount', palette='Set2')
# Set a y-limit to make the boxplot readable due to massive outliers in credit card data
plt.ylim(-10, 500) 
plt.title("Bivariate: Transaction Amount by Class (Zoomed in)")
plt.xlabel("Class (0 = Legitimate, 1 = Fraud)")
plt.ylabel("Transaction Amount ($)")
plt.show()
```
